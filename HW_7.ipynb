# Problem Statement:
311 data is an incredible resource, but one major concern with 311 complaints is that there can be bias in the data. The fact that one community district has a higher number of 311 noise complaints, doesn't necessarily mean it is actually louder than another community district. It simply means more people submitted 311 complaints. Some neighborhoods may have residents that are more likely to submit complaints than others for various reasons (ex: different expectations, more awareness about 311, more free time, more sense of empowerment or hope that the complaint will be resolved, etc). This makes it difficult to get a clear picture of conditions across the city, and it can help to compare datasets against each other.

Community District Profiles is an application that provides objective statistics about community districts. In this assignment, you will chose an indicator published on Community District Profiles and compare it against 311 complaint data.

```
# I always get 2 Unnamed rows, so I am dropping them
df = df.drop('Unnamed: 0', axis = 1)
df = df.drop('Unnamed: 0.1', axis = 1)
# making all complaint_types lowercase
df['complaint_type'] = df['complaint_type'].str.lower()
df.head()
```

```
# recoding borocd on original df dataframe

def recode_borocd_counts(row):
    if row.community_board[0] != 'U':
      if 'MANHATTAN' in row.community_board:
        return '1' + row.community_board[:2]
      elif 'BRONX' in row.community_board:
        return '2' + row.community_board[:2]
      elif 'BROOKLYN' in row.community_board:
        return '3' + row.community_board[:2]
      elif 'QUEENS' in row.community_board:
        return '4' + row.community_board[:2]
      elif 'STATEN ISLAND' in row.community_board:
        return '5' + row.community_board[:2]
    else:
      return 'Invalid borocd'

df['borocd'] = df.apply(recode_borocd_counts, axis=1)

# checking recode
df.head()
```

```
cddf_pov = cddf[['borocd','poverty_rate']]
cddf_pov.head()
```

# Step 2. 
Review the complaint types in the 311 complaint data. Choose the complaint type(s) you hypothesize should correlate with the indicator you chose in Step 2. Create a new dataframe that only captures the 311 complaints you want to analyze.

```
df = df[df.complaint_type == "homeless person assistance"]
df_comp = df[['borocd','complaint_type']].groupby(['borocd','complaint_type']).size().reset_index(name='complaints_by_cd')
df_comp.head()
```

# Step 3. 
Write out the hypothesis for your analysis. What relationship are you expecting to find between the variables and why?

**I expect to see a positive correlation between the variables.
Specifically, as poverty rate in the Indicators dataset increases, the complaints for homeless person assistance rate will increase
(because there will be more homeless people as the poverty rate increases).**

# Step 4. 
Get the count of 311 complaints per community district (borocd) using the filtered dataframe you created in Step 2.

```
df_comp['borocd'] = df_comp['borocd'].astype(int)
df_comp['complaints_by_cd'] = df_comp['complaints_by_cd'].astype(int)
df_comp.head()
```

# Step 5. 
Join the indicator dataframe (from Step 1) with 311 counts dataframe (from Step 4) using the borocd value.

```
homeless = pd.merge(left=df_comp, right=cddf_pov, left_on=['borocd'], right_on=['borocd'])
homeless.head()
```

# Step 6.
Produce a scatter plot of the indicator values vs. 311 complaint counts. The should be one point per community district.

*Note: I experimented with adding labels.*

```
plt.figure(figsize=(15, 6))

boro = homeless.borocd
x_coords = np.array(homeless.poverty_rate) # x
y_coords = np.array(homeless.complaints_by_cd) # y

for i,type in enumerate(boro):
    x = x_coords[i]
    y = y_coords[i]
    plt.scatter(x, y,)
    plt.text(x+0.4, y+0.4, type, fontsize=9)

plt.scatter(
    x_coords,
    y_coords,
    s=50, # size of marker
    alpha=0.5) # transparency. 0 is 100% transparent. 1 is 100% opaque

plt.title('Poverty Rate vs. 311 Homeless Person Assistance Complaints')
plt.xlabel('Community District Poverty Rate')
plt.ylabel('# of Homeless Person Assistance Requests per Community District')

plt.show()
```

# Step 7. 
Conduct an OLS linear regression using the indicator values vs. 311 complaint counts and interpret the results.

```
import statsmodels.api as sm 
```

```
# determine the correlation coefficient using the np.corrcoef() function from numpy

x = np.array(homeless.poverty_rate)
y = np.array(homeless.complaints_by_cd)

# produces a matrix of the correlation coefficients
np.corrcoef(x, y)
```

```
# Note that y is the first argument, before x. This is different than plotting and the correlation function.

# y = dependent variable (# of 311 requests)
y = np.array(homeless.complaints_by_cd)

# x = independent variable (population) with a constant added
x = np.array(homeless.poverty_rate)
x = sm.add_constant(x)

# print(x)

model = sm.OLS(y, x).fit()

# Print out the statistics

print(model.summary())
```

# Step 8. 
Use the linear regression results to add a line of best fit to your scatterplot

```
p = model.params # intercept and coefficient
print('parameters:', p)

r2 = model.rsquared
print('R2:', r2)
```

```
label = 'y = ' + str(round(p[1],2)) + 'x + ' + str(round(p[0],2)) + '; R2 = ' + str(round(r2,3))
print(label)
```

```
x_dummy = np.array(homeless.poverty_rate)

plt.figure(figsize=(15, 6))

plt.scatter(
    np.array(homeless.poverty_rate), # x
    np.array(homeless.complaints_by_cd), # y
    s=50,
    c='#3d03fc',
    alpha=0.5) 

plt.plot(
    x_dummy, 
    p[0] + p[1] * x_dummy,
    c='blue',
    label=label)

plt.title('Poverty Rate vs. 311 Homeless Person Assistance Complaints')
plt.xlabel('Community District Poverty Rate')
plt.ylabel('# of Homeless Person Assistance Requests per Community District')
plt.legend()

plt.show()
```
